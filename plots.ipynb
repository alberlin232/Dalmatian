```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./plots/OG.csv")
plt.plot(df["training_acc"], label="Training Accuracy")
plt.plot(df["val_acc"], label="Validation Accuracy")
plt.ylim((0,1))
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("WLASL100 SPOTER Poses")
plt.savefig("./plots/WLASL100_SPOTER.png", dpi=150)
```

```python
df = pd.read_csv("./plots/OG_adam.csv")
plt.plot(df["training_acc"], label="Training Accuracy")
plt.plot(df["val_acc"], label="Validation Accuracy")
plt.ylim((0,0.2))
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("WLASL100 SPOTER Poses with Adam")
plt.savefig("./plots/WLASL100_SPOTER_Adam.png", dpi=150)
```
